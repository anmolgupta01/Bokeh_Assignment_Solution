# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are the visual markers used to represent data points in a plot. They include various shapes (circles, squares, triangles, etc.), lines, patches, and more. Glyphs are added to a Bokeh plot to visualize your data in different ways. You can customize their appearance and attributes to create informative and visually appealing visualizations.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure
p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add glyphs to the plot
# Circles glyph
p.circle(x, y1, size=8, color="blue", legend_label="sin(x)")

# Squares glyph
p.square(x, y2, size=8, color="red", legend_label="cos(x)")

# Lines glyph
p.line(x, y1, line_width=2, color="blue", legend_label="sin(x)")
p.line(x, y2, line_width=2, color="red", legend_label="cos(x)")

# Display the plot
output_notebook()
show(p)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Customizing the appearance of a Bokeh plot involves modifying various attributes of different plot elements such as axes, title, legend, and more. 

Title and Axes:
You can customize the title and axis labels using properties of the figure object.

Legend:
You can customize the appearance of the legend using properties of the Legend object. Legends are associated with glyphs in the plot.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend

# Create a figure
p = figure(title="Customizing Legend", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add glyphs with legend labels
circle_glyph = p.circle(x=[1, 2, 3], y=[1, 2, 1], size=10, color="blue", legend_label="Circle")
line_glyph = p.line(x=[1, 2, 3], y=[3, 2, 3], line_width=2, color="red", legend_label="Line")

# Customize the legend
legend = Legend(items=[
    ("Circle", [circle_glyph]),
    ("Line", [line_glyph])
])
p.add_layout(legend, 'above')

# Display the plot
output_notebook()
show(p)
```
